```python
# Process tweet data
final_tweets = []
for x in tweets['tweets']:
    data = [
        x['link'],
        x['text'],
        x['date'],
        x['stats']['comments'],
        x['stats']['retweets'],
        x['stats']['quotes'],
        x['stats']['likes']
    ]
    final_tweets.append(data)

# Convert the list of lists into a DataFrame
tweets_df = pd.DataFrame(final_tweets, columns=[
    'URL',
    'Tweet Content',
    'Date and Time',
    'Comments Count',
    'Retweets Count',
    'Quotes Count',
    'Likes Count'
])

# Save the DataFrame to a CSV file
#csv_file_path = r'C:\Users\Downloads\Tweets.csv'
#tweets_df.to_csv(csv_file_path, index=False, encoding='utf-8')

print(tweets_df)
```
